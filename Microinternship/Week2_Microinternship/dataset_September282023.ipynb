```python
def total_opens():
    total_opens = df_emailopened["NumOpens"].sum()
    print("Total number of NumOpens: ", total_opens)



total_opens()
```

```python
average_opens = df_emailopened["NumOpens"].mean()
df_emailopened["AverageNumOpens"] = average_opens

print("Average number of NumOpens:", average_opens)


```

```python
minimum_opens = df_emailopened["NumOpens"].min()
df_emailopened["MinimumOpens"] = minimum_opens
print("Minimum number of NumOpens: ", minimum_opens)
```

```python
maximum_opens = df_emailopened["NumOpens"].max()
df_emailopened["MaximumOpens"] = maximum_opens
print("Maximum number of NumOpens: ", maximum_opens)
```

```python
df_emailopened["DateOpened"] = pd.to_datetime(df_emailopened["DateOpened"], format="%d/%m/%Y %H:%M") #convert string to datetime
```

```python
df_emailopened["DateOpenedLast"] = pd.to_datetime(df_emailopened["DateOpenedLast"], format="%d/%m/%Y %H:%M") #convert string to datetime
```

```python
df_emailopened["EngagementSpan"] = df_emailopened["DateOpenedLast"] - df_emailopened["DateOpened"]
df_emailopened.head(10)

```

```python
df_emailopened["EngagementHours"] = df_emailopened["EngagementSpan"].dt.total_seconds() / 3600
df_emailopened
```

```python
def numopen_level_classification(opens): #categorising open levels into different categories
    if opens <= 1:
        return 'Low'
    elif opens <= 4:
        return 'Medium'
    else:
        return 'High'
    
    
```

```python
df_emailopened["Openlevel"] = df_emailopened["NumOpens"].apply(numopen_level_classification)
df_emailopened
```

```python
df_emailopened["Openlevel"].value_counts().plot(kind='bar', title='Email Engagement Levels')
plt.ylabel("Number of People")
plt.show()
df_emailopened["Openlevel"].value_counts()
#high levels of medium open level
#low levels of high open level
```

```python
df_emailopened.groupby("EmailType")["AverageNumOpens"].mean().plot(kind="bar", title="Average Opens per Email Type")
plt.ylabel("Average Opens")
plt.show()
df_emailopened.groupby("EmailType")["AverageNumOpens"].value_counts()
```

```python
bins = pd.cut(df_emailopened['EngagementHours'], bins=10).value_counts().sort_index()
print(bins)

df_emailopened['EngagementHours'].plot(kind='hist', bins=10, title='Distribution of Engagement Hours')
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(
    x='Openlevel',
    y='EngagementHours',
    data=df_emailopened,
    boxprops=dict(facecolor='white', edgecolor='black'),
    whiskerprops=dict(color='black'),
    capprops=dict(color='black'),
    medianprops=dict(color='black'),
    flierprops=dict(marker='o', color='black', alpha=0.3, markersize=3),
    showfliers=True
)

plt.yscale('log')  #log scale to handle skewed data
plt.title('Engagement Span by Open Level (Log Scale)')
plt.grid(True, which="both", axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()

sns.boxplot(x='Openlevel', y='EngagementHours', data=df_emailopened).set_title('Engagement Span by Open Level')


df_emailopened.groupby("Openlevel")["EngagementHours"].std()
```
